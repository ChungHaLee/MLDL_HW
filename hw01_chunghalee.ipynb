```python
# Q5-e. Using the full data set, investigate the predictors graphically, using scatterplots or other tools of your choice. Create some plots highlighting the relationships among the predictors. Comment on your findings
```

```python
plt.plot('acceleration',  # x

         'weight',  # y

         data=auto_dataset, 

         linestyle='none', 

         marker='o', 

         markersize=10,

         color='blue', 

         alpha=0.5)

plt.title('Scatter Plot', fontsize=20)

plt.xlabel('acceleration', fontsize=14)

plt.ylabel('weight', fontsize=14)

plt.show()
```
